EXPLORATORY DATA ANALYSIS ON DIWALI SALES

IMPORTING NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

UPLOADING DATASET

```python
data = pd.read_csv(r'C://Users//NUHA//Desktop//python//Python_Diwali_Sales_Analysis-main//Diwali Sales Data.csv', encoding='unicode_escape')
```

DATA CLEANING

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data
```

```python
data['Amount'] = data['Amount'].astype('int')
```

```python
data['Amount'].dtypes
```

```python
data.columns
```

```python
data[['Age',"Orders","Amount"]].describe()
```

EXPLORATORY DATA ANALYSIS

GENDER

```python
data.columns
```

```python
ax= sns.countplot(x='Gender',data = data)

sns.set(rc={'figure.figsize':(1,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gender=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(2,5)})
sns.barplot(x='Gender',y = 'Amount',data=sales_gender)
```

AGE

```python
ax= sns.countplot(x='Age Group',data = data, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Age=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y = 'Amount',data=sales_Age)
```

STATES



```python
sales_States=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='State',y = 'Orders',data=sales_States)
```

```python
sales_States_Amount=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='State',y = 'Amount',data=sales_States_Amount)
```

MARITAL STATUS

```python
ax= sns.countplot(x='Marital_Status',data = data)
sns.set(rc={'figure.figsize':(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y = 'Amount',data=sales_marital,hue='Gender')
```

OCCUPATION

```python
ax= sns.countplot(x='Occupation',data = data)
sns.set(rc={'figure.figsize':(18,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Ocption=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='Occupation',y = 'Amount',data=sales_Ocption)
```

PRODUCT CATOEGORY


```python
ax= sns.countplot(x='Product_Category',data = data)
sns.set(rc={'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y = 'Amount',data=sales_prod)
```

PRODUCT ID

```python
data.columns
```

```python
sales_prod=data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(x='Product_ID',y = 'Orders',data=sales_prod)
```
